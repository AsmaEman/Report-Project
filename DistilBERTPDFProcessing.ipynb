```python
# !pip install pdfplumber
# !pip install transformers
# !pip install PyMuPDF

import pdfplumber
from transformers import pipeline

import fitz  # PyMuPDF
import io

def extract_text_from_pdf(pdf_path, start_page=None, end_page=None):
    with open(pdf_path, 'rb') as file:
        pdf_data = file.read()
    pdf_document = fitz.open("pdf", pdf_data)
    text = ""

    # Determine page range
    start_page = start_page or 0
    end_page = end_page or pdf_document.page_count - 1

    for page_num in range(start_page, min(end_page + 1, pdf_document.page_count)):
        page = pdf_document[page_num]
        text += page.get_text()

    return text

def search_in_pdf(question, pdf_path):
    pdf_text = extract_text_from_pdf(pdf_path, start_page=0, end_page=5)  # Adjust page range as needed
    qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad", tokenizer="distilbert-base-cased-distilled-squad")
    result = qa_pipeline(question=question, context=pdf_text)
    return result['answer']

# Example usage
pdf_path = '/content/sample_data/calcrim_2023_edition.pdf'
text_from_pdf = extract_text_from_pdf(pdf_path, start_page=0, end_page=5)


question = "What will be penal code for theft?"
answer = search_in_pdf(question, pdf_path)
print("Answer:", answer)
```
